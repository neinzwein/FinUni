```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.utils import to_categorical
```

```python
numpy.random.seed(7)
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
char_to_int = dict((c, i) for i, c in enumerate(alphabet))
int_to_char = dict((i, c) for i, c in enumerate(alphabet))
```

```python
seq_length = 3
dataX = []
dataY = []
```

```python
for i in range(0, len(alphabet) - seq_length, 1):
    seq_in = alphabet[i:i + seq_length]
    seq_out = alphabet[i + seq_length]
    dataX.append([char_to_int[char] for char in seq_in])
    dataY.append(char_to_int[seq_out])
    print(f'{seq_in} -> {seq_out}')
```

```python
X = numpy.reshape(dataX, (len(dataX), 1, seq_length)) 
X = X / float(len(alphabet))
y = to_categorical(dataY)
```

```python
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(y.shape[1], activation='softmax'))
```

```python
#метод стохастического градиентного спуска
#метод стохастического градиентного спуска, основанный на адаптивной оценке моментов первого и второго порядка 80
#метод стохастического градиентного спуска, который использует оператор знака для управления величиной обновления, в отличие от других адаптивных оптимизаторов, таких как Adam, которые полагаются на моменты второго порядка.
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=500, batch_size=1, verbose=2)
```

```python
scores = model.evaluate(X, y, verbose=0)
print("Model Accuracy: %.2f%%" % (scores[1]*100))
```

```python
for pattern in dataX:
    x = numpy.reshape(pattern, (1, 1, len(pattern)))
    x = x / float(len(alphabet))
    prediction = model.predict(x, verbose=0)
    index = numpy.argmax(prediction)
    result = int_to_char[index]
    seq_in = [int_to_char[value] for value in pattern]
    print(seq_in, "->", result)
```
