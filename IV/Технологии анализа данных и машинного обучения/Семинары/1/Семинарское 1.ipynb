```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Степень принадлежности 
def trapezoid_check(x,a=1,b=3,c=5,d=7):
    if x <= a or x >= d:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return 1
    elif c < x < d:
        return (d - x) / (d - c)
```

```python
def Function(subset,a,b,c,d):
    res = []
    for x in subset:
        res_x =  trapezoid_check(x, a, b, c, d)
        res.append((x, res_x))
    return res
```

```python
#ЗАДАЮТСЯ ЗНАЧЕНИЯ СЮДА (ЧЕТКИЕ ОБЪЕКТЫ МНОЖЕСТВА)
subset=[i for i in range(10)]
```

```python
#(ПО УМОЛЧАНИЮ ПРОСТАВЛЕНЫ А=1 и так далее, нужно поменять, если нужно - 'это сама трапеция')
res_set = Function(subset, a=1, b=3, c=5, d=7)
for x, res_x in res_set:
    print("Элемент четкого множества: "+ str(x) +" , Дополнение нечеткого множества: " + str(res_x))
```
