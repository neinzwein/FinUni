```python
import matplotlib.pyplot as plt
```

```python
N =[25,20,20,15,10,9,1,0]
V = [5,4,2,4,2,2,1,0]
res=[]
Error_total = 110

for i in range(len(N)-1):
    res.append(((Error_total-N[i])/(Error_total-V[i]))*(N[i+1]+V[i+1]))

run_numbers = list(range(1, len(N)))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(run_numbers, res, marker='o', linestyle='-', color='b')
plt.xlabel(' ')
plt.ylabel('   ')
plt.title('      ')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score
true_values = N[1:]

r2 = r2_score(true_values, res)

print(f"R: {r2}") #   
```
